```python
import pandas as pd
```

```python
# cria um dataframe

datas = {'nome': ['Argentina', 'Brasil', 'França', 'Italia', 'Reino unido'],
         'continente': ['America', 'America', ' Europa', 'Europa', 'Europa'],
         'extensao': [2780, 8511, 644, 301, 244],
         'corVerde': [0, 1, 0, 1, 0]
         }

codes = ['ARP', 'BR', 'FR', 'IT', 'UK']

countries = pd.DataFrame(datas, index=codes)

print(countries)

countries
```

```python
# propriedades basicas dos DataFrames

num_lines = countries.shape[0]
num_columns = countries.shape[1]
position = countries.index
columns = countries.columns
countries_type = type(countries)
countries_type = countries.dtypes
countries_idx_dtype = countries.index.dtype

print('numero de linhas: ', num_lines)
print('numero de colunas: ', num_columns)
print('rotulo das linhas: ', position)
print('rotulos das colunas: ', columns)
print('tipo (type): ', countries_type)
print('dtype das colunas: \n', countries_type)
print('dtype dos rotulos das linhas: ', countries_idx_dtype)
```

```python
# buscas no dataframme

# cria o dataframe

datas = {'nome': ['Argentina', 'Brasil', 'França', 'Italia', 'Reino unido'],
         'continente': ['America', 'America', ' Europa', 'Europa', 'Europa'],
         'extensao': [2780, 8511, 644, 301, 244],
         'corVerde': [0, 1, 0, 1, 0]
         }

codes = ['ARP', 'BR', 'FR', 'IT', 'UK']

countries = pd.DataFrame(datas, index=codes)

# testa se um dado rótulo de linha existe

has_BR = 'BR' in countries.index
has_US = 'US' in countries.index
print("existe o rótulo 'BR'? -> ", has_BR)
print("existe o rótulo 'US'? -> ", has_US)
print("_________________________________")

# testa se um dado rótulo de coluna existe

has_greenColor = 'corVerde'in countries.columns
has_blueColor = 'corAzul' in countries.columns
print("existe a coluna 'corVerde'? -> ", has_greenColor)
print("existe a coluna 'corAzul'? -> ", has_blueColor)
print("_________________________________")

# testa se o valor faz parte de uma coluna

has_Brasil = countries['nome'].isin(['Brasil'])
print("existe o valor 'Brasil' na coluna 'nome'?")
print(has_Brasil)
```

```python
# modificando o dataframe

print("--- dataframe original ---")
print(countries)

# inserindo o país japao

countries.loc['JP'] = {
    'nome': 'Japao',
    'continente': 'Asia',
    'extensao': 372,
    'corVerde': 0
}

print("dataframe modificado")
print(countries)

# altera extensao brasil

countries.at['BR', 'extensao'] = 8512
print("dataframe modificado")
print(countries)

# remove argentina e reino unido

countries = countries.drop(['ARP', 'UK'])
print("dataframe modificado")
print(countries)
```
